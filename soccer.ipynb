```python
df1 = pd.DataFrame(data_frame, columns = ['home_team','home_team_fifa_rank','away_team','away_team_fifa_rank'])
df1
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x = 'home_team_fifa_rank',y = 'away_team_fifa_rank',data = df1)
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[("index","$index"),("(home_team_fifa_rank, away_team_fifa_rank)","($home_team_fifa_rank, $away_team_fifa_rank)")])
p = figure(title='simple line example',x_axis_label = 'home_team_fifa_rank',y_axis_label = 'away_team_fifa_rank', plot_width = 1000,plot_height = 1000, tools=[TOOLTIPS])
p.circle('home_team_fifa_rank','away_team_fifa_rank',size = 10, source = df1)
show(p)
```
